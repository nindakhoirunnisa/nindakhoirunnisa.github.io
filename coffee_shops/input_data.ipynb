### Outlet Types

```python
outlets = pd.read_csv('sales_outlet.csv')
outlets
```

It is known that there are two outlet types.

```python
staffs = pd.read_csv('staff.csv')
store_types = staffs['location'].unique()
store_types
```

However, in `staff` dataset, there are 4 types of outlet: retail (3, 4, 5, 6, 7, 8, 9, 10), WH (warehouse), HQ (headquarter), and FL (it is assumed to be flagship store). 
Type 'retail' and 'warehouse' are already in `sales_outlet` dataset meanwhile 'flagship store' and 'headquarter' are not. Thus, mapping is needed to turn those data into same format.

```python
code_map = {
    'HQ': 'headquarter',
    'WH': 'warehouse',
    'FL': 'flagship store'
}

# Map each code to its description or 'retail' if not found in the dictionary
outlet_types = [code_map[type] for type in store_types if type in code_map]

# Add 'retail' if it hasn't been added yet
if 'retail' not in outlet_types:
    outlet_types.append('retail')

print(outlet_types)
```

```python
# Insert the data to SQL table
try:
    cursor = connection.cursor()
    for outlet_type in outlet_types:
        sql = "INSERT INTO OUTLET_TYPES (name) VALUES (%s)"
        cursor.execute(sql, (outlet_type,))
        connection.commit()
    print("Data inserted successfully")
except mysql.connector.Error as error:
    print("Failed to insert data into MySQL table:", error)
```
